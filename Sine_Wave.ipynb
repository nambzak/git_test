```python
import numpy as np
import matplotlib.pyplot as plt

amp = 2 #Acting as our Amplitude

freq = 2 #Acting as frequency

phi = 0 # Acting as the phase

sr = 1000 # number of data points/samples in 1 sec

time = np.arange(0, 10, 1 / sr) # time axis

y =  amp * np.sin(2 * np.pi * freq * time + phi)  # equation to get a sine wave is this

plt.figure(figsize = (10, 4))  # size of our plot

plt.plot(time, y)

plt.title("A SINE WAVE SIMULATION USING PYTHON")

plt.xlabel("TIME(s)")

plt.ylabel("AMPLITUDE")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Below is another Example of a sine wave

# Set up parameters

n = 1000  # Number of data points

t = np.linspace(0, 2*np.pi, n)  # Time vector to complete 1 cycle

freq = 2  # Frequency of sine wave
amp = 4  # Amplitude of sine wave

# Generate sine wave
signal = amp * np.sin(2*np.pi*freq*t)

# Plot signal
plt.figure(figsize=(10, 4))
plt.plot(t, signal)
plt.xlabel('TIME (s)')
plt.ylabel('AMPLITUDE')
plt.title('SINE WAVE SIMULATION')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#NOISE ADDED INTO A SIGNAL
import numpy as np
import matplotlib.pyplot as plt

#ADDING GAUSSIAN NOISE INTO OUR SINE WAVE

n = 1000  # DATA POINTS

t = np.linspace(0, 10*np.pi, n)  # Time vector

freq = 2  # Frequency of sine wave

amp = 2  # Amplitude of sine wave

noise = 0.1  # Amplitude of Gaussian noise

# Generate sine wave
signal = amp * np.sin(2*np.pi*freq*t)

# Generate Gaussian noise
gaussian_noise = np.random.normal(0, noise, n)

# Add noise to signal
noisy_signal = signal + gaussian_noise

# Plot signal and noisy signal
plt.figure(figsize=(10, 4))
plt.plot(t, signal, label='SIGNAL')
plt.plot(t, noisy_signal, label='NOISY SIGNAL')
plt.title("GAUSSIAN NOISE ADDED TO THE SINE WAVE")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
import time

#ANOTHER EXAMPLE OF GAUSSIAN NOISE ADDED TO A SINE WAVE
freq_sample = 800 #signal read 800 times in a second

freq = 5 #number of cycles in a sec

samples = 700 # samples start from 0 to 700 dsitributed evenly

x_input = np.arange(samples) #this equates to time vector

noise = 0.0008 * np.asarray(random.sample(range(0, 1000), samples)) # CREATE RANDOM MAGNITUDES TO REP NOISE PUT IN AN ARRAY SHAPE

y_out = np.sin( 2 * np.pi * freq * x_input / freq_sample) + noise

plt.figure(figsize=(10, 4))
plt.title("GAUSSIAN NOISE ADDED TO A SINE WAVE")
plt.plot(x_input, y_out)
plt.show()
```
